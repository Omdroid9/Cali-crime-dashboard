Q11
Create the following graph using the following properties

We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object, or axis('tight') for automatically getting "tightly fitted" axes ranges:

```python

x = np.linspace(0, 5, 100)
y1 = x ** 2
y2 = x ** 3

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
fig.suptitle("multiple plots")  

axes[0].plot(x, y1, label="x**2")
axes[0].plot(x, y2, label="x**3")
axes[0].set_title("default axes ranges")
axes[0].legend()  
axes[1].plot(x, y1, label="x**2")
axes[1].plot(x, y2, label="x**3")
axes[1].set_title("tight axes")
axes[1].axis('tight')   
axes[1].legend()

axes[2].plot(x, y1, label="x**2")
axes[2].plot(x, y2, label="x**3")
axes[2].set_title("custom axes range")
axes[2].set_xlim(2, 5)    
axes[2].set_ylim(0, 60)  
axes[2].legend()

plt.tight_layout()  
plt.show()
```

Q12 Create the following using data in x and y 

```python


x_main = np.linspace(0, 5, 100)
y_main = x_main ** 2

x_inset = np.linspace(0, 25, 100)
y_inset = np.sqrt(x_inset)

fig = plt.figure(figsize=(6,4))
ax_main = fig.add_axes([0.1, 0.1, 0.8, 0.8])  

ax_main.plot(x_main, y_main, 'b')
ax_main.set_xlabel("X_label_axes2")
ax_main.set_ylabel("Y_label_axes2")
ax_main.set_title("Axes 2 Title")

ax_inset = fig.add_axes([0.2, 0.5, 0.4, 0.3])  

ax_inset.plot(x_inset, y_inset, 'r')
ax_inset.set_xlabel("X_label_axes2")
ax_inset.set_ylabel("Y_label_axes2")
ax_inset.set_title("Axes 2 Title")

plt.show()
```
